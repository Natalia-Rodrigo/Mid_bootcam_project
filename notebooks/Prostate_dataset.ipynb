**Load DataSet from CSV with Pandas**

```python
DataSet1 = '../Raw_Data/Prostate_GSE46602.csv'
DataSet2 = '../Raw_Data/Prostate_GSE46602.csv'

main = pd.read_csv(DataSet1)
display(main.head()), display(main.tail())
```

**Perform a Quantile Transformation**

Quantile transformation is applied to gene expression data to normalize its distribution, addressing issues of skewness and outliers. By reshaping the data to adhere to a standard normal distribution, quantile transformation ensures statistical consistency, mitigates the impact of outliers, and improves the performance of certain analyses and machine learning models that assume normally distributed input. This preprocessing step enhances interpretability and facilitates valid comparisons across different datasets.

```python
# Perform the QuantileTransformer and save for later use
main_transformed, qt = cf.quantileTransformer(main)
main_transformed.head()
```

**Explore the Quantile Transformation set**

```python
# Create two melted dataframs to explore within graphs
transformed_melt = main_transformed.melt(id_vars=['samples', 'type'], var_name='gene', value_name='expression').sort_values('type')
main_melt = main.melt(id_vars=['samples', 'type'], var_name='gene', value_name='expression').sort_values('type')

# Filter Melts on 6 samples
qt_set = transformed_melt.copy()
qt_set = qt_set[qt_set['samples'].isin([185, 184, 183, 160, 161, 162])]

main_set = main_melt.copy()
main_set = main_set[main_set['samples'].isin([185, 184, 183, 160, 161, 162])]

# plot log transformed data on boxplot
f, ax = plt.subplots(1,2,figsize=(10, 5), sharey=True, sharex=True)
ax[1] = sns.boxplot(x='samples', y='expression', hue = 'type',  data=qt_set, width = 0.3, ax=ax[1])
ax[0] = sns.boxplot(x='samples', y='expression', hue = 'type',  data=main_set, width = 0.3, ax=ax[0])

# Set titles, Labels and legendas
ax[0].legend([])
ax[1].legend(loc='upper left')

ax[1].set_title('Quantile Transformed')
ax[0].set_title('Non-Quantile Transformed')

ax[0].set_xlabel('Samples')
ax[1].set_xlabel('Samples')
ax[0].set_ylabel('Expression')

# Show the plot
plt.tight_layout()
plt.show()
```

**Perform a two sample T Test**

A two-sample t-test is employed to assess whether there is a significant difference in the mean expression levels between tumoral and normal genes. This statistical test is applied to identify genes that exhibit a statistically significant impact on the tumoral set compared to the normal set. By isolating genes with significant differences, researchers can focus on those likely to play a role in tumorigenesis, aiding in the identification of potential biomarkers or therapeutic targets.

```python
# Get list of ExpressionGenes with the cleaning function
expression_genes = cf.getListDifferentialGenes(df=main_transformed, psig=0.01, tumor_column='type', tumor_value='tumoral')
print(len(expression_genes))
```

Create new set from Expression genes list

```python
# Create new set to only include the expression_Genes from previous function
main_transformed = main_transformed[['samples', 'type'] + expression_genes]
main_transformed.head()
```

**Checking and filtering for multicollinearity genes**

Filtering genes with high correlation is a crucial step to reduce overfitting in the final model. High correlation between genes can lead to multicollinearity, where redundant information is present, potentially causing the model to be overly sensitive to the training data. By excluding correlated genes, the model becomes more robust, enhancing its generalization performance and improving its ability to capture meaningful patterns in the gene expression data.

```python
# Find correlated genes, and create exclude list
correlated_genes_list, exclude_list, r_values = cf.find_correlated_genes(main_transformed.drop('type', axis=1))
print(len(exclude_list))
display(r_values.head())
```

```python
# Filter expression_gene list for exclude list from previous function
final_gene_list = []

for col in expression_genes:
    if (col not in exclude_list):
        final_gene_list.append(col)

print(len(final_gene_list))
```

```python
# Create final filtered df
final_set = main_transformed[['samples', 'type'] + final_gene_list]
final_set.head()
```

**Train Model**

A logistic regression model was employed to predict whether a sample belongs to the normal or tumoral category based on the filtered gene set. Logistic regression is suitable for binary classification tasks, making it apt for distinguishing between normal and tumoral samples. By leveraging the selected genes and their expression patterns, the logistic regression model provides a probabilistic prediction, aiding in the identification and classification of samples into the respective categories.

```python
LogisticReg, cm, scores = cf.trainLogisticModel(final_set, ColumnToPredict='type', tumor_value='tumoral')

print('The scores of the model are:')
for key in scores.keys():
    print(f"{key} : {round(scores[key],2)} ")
```

**Save Models**

Save all models used within this notebook

```python
import os
import pickle

def savemodel(model, path, filename):
    isExist = os.path.exists(path)
    if not isExist:
        os.makedirs(path)
        print("The new directory is created!")

    with open(path+filename, "wb") as file:
        pickle.dump(model, file)


savemodel(qt, '../Transformer/', 'tim_Prostate_GSE46602.pkl')
savemodel(LogisticReg, '../Models/', 'tim_Prostate_GSE46602.pkl')
```
